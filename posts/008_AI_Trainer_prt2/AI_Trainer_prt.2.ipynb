## **Introduction** 

---

In my first post, I asked an LLM to design a two-week training plan for my touch rugby camp. The plan was well structured and covered everything from conditioning to ball-handling, but it was also built without any real understanding of who I am as an athlete. It didn’t ask about my schedule, my position, my strengths and weaknesses, or whether I had any injuries. A real coach wouldn’t design a program without asking those things first.

For this follow-up post, I wanted to see whether I could make the AI behave more like an actual coach by changing the prompt. I used a more structured approach based on prompting techniques that focus on giving the AI a defined role, a clear process, and the instruction to ask questions before producing anything. This mirrors the way coaches gather information before building a training plan.


### How I Rewrote the Prompt

To redesign the prompt, I followed ideas that come from prompt experts like Ethan Mollick and from general research on structured prompting. Instead of asking the AI for a plan directly, I broke the request into clear phases that a real coach would follow.

First, I gave the model a defined role by telling it to act as a high-performance touch rugby coach. This helps the AI adjust its tone, structure, and level of detail.

Second, I required a one-question-at-a-time intake process. This forces the AI to personalize the plan and prevents it from skipping to generic answers.

Third, I told it which specific categories it needed to ask about, including conditioning, sprint speed, ball-handling ability, injury history, available training time, and equipment access. Without this step, AI tends to guess or assume.

Fourth, I asked it to summarize its understanding of me before creating the plan. This clarifies how well it interpreted my answers.

Finally, I required the training plan to directly reflect my responses. This prevents the AI from giving me a random template and makes it build something that actually fits my life.

By structuring the prompt this way, I was essentially teaching the AI to behave less like a search engine and more like an actual coach who listens first and designs second. The next section shows the exact prompt I used.

## **The Improved Prompt:**
> You are a high-performance touch rugby coach helping me prepare for a three-day camp from eight in the morning to four in the afternoon. Follow this process.
First, ask me one question at a time to collect at least ten specific details about my fitness level, sprint ability, conditioning base, ball-handling level, touch rugby experience and position, injury history, weekly schedule, recovery habits, equipment access, and goals.
Second, summarize your understanding of me as an athlete.
Third, create a fully personalized two-week training plan that directly reflects my answers. The plan should include conditioning, agility, sprint work, ball-handling, strength and stability, recovery practices, nutrition, hydration, and mental preparation for decision-making and game awareness.
Fourth, ask if I want adjustments.
Do not skip steps.
Do not begin the plan until your assessment is complete.


## **What the AI Asked Me**

To keep everything clean, each question is formatted as a quote.

> **Question 1:**
> What is your current fitness level — for example, how many times per week do you work out, and what types of training are you doing right now?

> **Question 2:**
> How would you describe your sprinting ability right now? Are you explosive off the mark? Do you sustain high effort over short distances? How often do you do sprint work?

> **Question 3:**
> What is your current conditioning base like? Can you handle repeated high-intensity efforts?

> **Question 4:**
> How would you rate your current ball-handling skills (passing accuracy, catching under pressure, speed of execution)?

> **Question 5:**
> How much experience do you have in touch rugby? What position do you usually play?

> **Question 6:**
> Do you have any current or past injuries affecting you?

> **Question 7:**
> What does your typical weekly schedule look like? How many days can you realistically train?

> **Question 8:**
> What do your recovery habits look like? Sleep, stretching, mobility, hydration?

> **Question 9:**
> What equipment and facilities do you have access to?

> **Question 10:**
> What are your specific goals for this camp?

## **AI’s Summary of Me as an Athlete**

```markdown
You’ve played touch rugby for 7–9 years, primarily at middle or link. You’re explosive off the mark and skilled, but your conditioning and sprint repeatability need work. Passing accuracy and speed are areas you want to improve. Past injuries include an ACL tear, a hamstring strain, and ankle instability, so joint stability and smart recovery are priorities.

You currently work out 2–4 times per week and can commit 3–5 sessions in the next two weeks. You have access to a gym and a grass field. You sleep well but want to improve mobility and recovery practices. Your goals include improving fitness, sharpening skills, boosting confidence, and standing out at camp.
```

## **The New Training Plan**
### **Overview**

**Weekly Commitment:** 3–5 sessions
**Session Length:** 60–90 minutes
**Targets:** Conditioning base, anaerobic repeatability, agility, sprint exposure, passing, joint stability, core strength, recovery, hydration, resilience

---

# **WEEK 1**

## **Day 1 — Field + Skill (75 min)**

**Goal:** Speed exposure, lateral movement, fundamental passing
• Dynamic warm-up
• Sprint mechanics + agility
• Passing under load
• Conditioning intervals
• Cooldown
**Nutrition:** 20–30g protein post
**Hydration:** 2–3 L/day + electrolytes

---

## **Day 2 — Gym: Lower + Stability (75 min)**

**Goal:** Glutes, hamstrings, ankle/knee stability
• Warm-up
• Strength block
• Stability and core
• Mobility cooldown
**Sleep:** 7.5 hours minimum

---

## **Day 3 — Field: Anaerobic Repeat + Passing (70 min)**

• Warm-up
• Repeat sprint conditioning
• Ball-handling flow
• Cooldown
**Mental Prep:** Visualize camp scenarios

---

## **Day 4 — Active Recovery (Optional)**

Light jog, cycle, foam roll, mobility, breathing reset

---

## **Day 5 — Gym: Upper/Total Body**

Push–pull circuits, med ball work, shoulder stability

---

## **Day 6 — Game Prep / Positional Focus**

3v3 touch scenarios, defense resets, dump options, agility reps

---

## **Day 7 — Off or Mobility Flow**

Easy day, stretch, hydrate, refuel

---

# **WEEK 2 — “Specificity + Camp Sharpening”**

## **Day 1 — Field**

Touch simulation, sprint fade passes, tempo play, mini games

---

## **Day 2 — Gym: Power**

Jump squats, cleans (if confident), Nordic curls, ankle/neck prehab

---

## **Day 3 — Lactate Session (Field)**

30s shuttles with 15s rest
Game-vision drills

---

## **Day 4 — Recovery Day**

Sauna or contrast
Deep stretching

---

## **Day 5 — Final Touch Sharpener**

Warm-up, shadow defense, passing under fatigue, mini continuous games

---

## **Day 6 — Off Day**

Hydrate, eat well, early sleep

---

## **Day 7 — Mental Rehearsal**

Light movement, 3 sprints at 70 percent
Visualization and confidence cues

## **Reflection:**

Comparing this plan to the one from my first post, the main difference is how specific it is. The earlier plan covered everything a typical touch rugby program would include, but it was clearly built without much understanding of me as an athlete. It did not account for my position, my conditioning level, or my injury history. It worked as a general outline, but not something I would realistically follow as a full training schedule.

With the new prompt, the AI took the time to ask questions first, and that made the final plan feel more relevant. The conditioning matches where I am right now. The stability work makes sense for my ACL and ankle history. The skill work reflects what I want to improve. The weekly structure also fits the amount of time I can actually train.

There are still parts that feel overstructured, and it is not the same as working with a real coach who adjusts things based on how I feel that day. Even so, this version is more useful and more realistic. The improvement came from giving clearer instructions and requiring the AI to gather information before creating anything. If I use AI again for training ideas, this is the approach I would use.
