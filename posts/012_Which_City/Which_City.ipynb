## **Introduction:**

Growing up in Taipei spoiled me in the best way. Everything was easy. Food at every hour. Public transit that actually works. Streets that feel alive but never overwhelming. When I moved to Philadelphia for college, it honestly felt like learning how to live all over again. Suddenly I had to think about safety, transportation, groceries, weather, everything. I didn’t realize how much a city shapes your daily life until I had to start rebuilding those routines from scratch.

Now that I’m in my early twenties and about to graduate, I keep wondering what comes next. Where am I supposed to go after this? What city actually makes sense for the kind of adult life I want? I still don’t fully know what that looks like, but I do know that choosing where to live feels like choosing a version of myself.

So for fun, and maybe also out of mild panic, I decided to ask an LLM to rank cities around the world for someone in their early twenties trying to figure things out. I wanted to see what it would pick based on things like affordability, community, nightlife, culture, and general vibes. Not because AI has all the answers, but because I’m curious whether it can help me think about this in a new way. Maybe it’ll give me ideas. Maybe it’ll tell me something I already knew. Either way, it feels like a pretty entertaining experiment while I try to sort out my next steps.

## **Prompt:** 

> I’m in my early twenties, about to graduate, and trying to figure out where I should live next. I grew up in Taipei and currently study in Philadelphia, so I’m used to cities with very different lifestyles. Can you rank around 10 cities around the world that would be great for someone in my stage of life?

> Consider things like affordability, job opportunities, social life, walkability, public transportation, general safety, community feel, dating and friendships, culture, food, and overall quality of life for young people.

> For each city, give a short explanation of why it might be a good fit.
Also tell me which cities might match someone who grew up in Taipei, and which feel closer to Philadelphia.

> If you need more details about what I want or my personality, ask me first.

## **Response:**

> This is a fantastic question, and your background in both Taipei and Philadelphia gives you a unique perspective!
> You've experienced very different urban environments, which will help you appreciate the nuances of other cities.
> I'll rank around 10 cities that offer a great balance for someone in their early twenties, focusing on the criteria you mentioned.
> "Affordability" is always relative, especially in global cities, so I'll consider it in comparison to other major hubs.

> ---
> ### Top Cities for Early-Twenties Graduates
> **1. Berlin, Germany**
> Why it's a great fit: Berlin is a magnet for young people due to its unique blend of history, vibrant arts scene, and burgeoning tech industry.
> It's relatively affordable compared to other major European capitals, especially for rent.
> The social scene is incredibly diverse, with countless bars, clubs, cafes, and cultural events.
> Public transport is excellent, and the city is very bike-friendly.
> There's a strong international community, making it easier to meet people.
> Job opportunities are growing, particularly in startups and tech.
> The food scene is varied and affordable, and safety is generally high.
> Key Strengths: Culture, affordability (relative), social life, public transport, job opportunities (tech).
>
> **2. Lisbon, Portugal**
> Why it's a great fit: Lisbon has exploded in popularity, and for good reason.
> It offers an incredibly high quality of life at a more affordable price point than many Western European cities.
> The city is stunning, with a vibrant social scene, delicious food, and a warm climate.
> There's a growing tech and startup ecosystem, attracting many young professionals.
> Walkability is high (though hilly!), and public transport is good.
> The community feel is welcoming, and it's easy to meet people, especially with the influx of expats.
> Dating and friendships thrive in its relaxed atmosphere.
> Key Strengths: Affordability (relative), social life, culture, food, climate, growing job market.
>
> **3. Toronto, Canada**
> Why it's a great fit: Toronto is a truly multicultural metropolis with a robust economy and diverse job opportunities across many sectors (finance, tech, media, healthcare).
> It's incredibly safe, has excellent public transportation, and boasts a vibrant social scene with endless options for food, nightlife, and culture.
> While not cheap, it offers a higher earning potential than some European cities.
> It's very walkable in many neighborhoods, and the community is highly welcoming to newcomers and international residents.
> Key Strengths: Job opportunities, diversity, safety, public transport, social life, culture.
>
> **4. Melbourne, Australia**
> Why it's a great fit: Consistently ranked among the world's most livable cities, Melbourne offers an exceptional quality of life.
> It's renowned for its arts, culture, food, and coffee scene.
> Job opportunities are strong, especially in finance, healthcare, and tech.
> The city is very walkable, with an extensive tram network.
> It has a lively social scene, great community feel, and a strong emphasis on outdoor activities.
> While rent can be high, salaries are generally competitive.
> Key Strengths: Quality of life, culture, food, job opportunities, safety, public transport.
>
> **5. Amsterdam, Netherlands**
> Why it's a great fit: Amsterdam offers a fantastic quality of life, especially for young people.
> It's incredibly international, with a large expat community, making it easy to meet people.
> The city is beautiful, very safe, and highly walkable/bikeable with excellent public transport.
> Job opportunities are good, particularly in tech, creative industries, and international business, and many companies operate in English.
> The social scene is vibrant, and there's a strong emphasis on work-life balance.
> The main drawback is the high cost of living, especially for housing.
> Key Strengths: Quality of life, international community, safety, walkability/biking, job opportunities, culture.
>
> **6. Chicago, USA**
> Why it's a great fit: For a major US city, Chicago offers a surprisingly good balance of job opportunities, culture, and relative affordability (compared to NYC or SF).
> It's a major hub for finance, tech, healthcare, and manufacturing.
> The city has stunning architecture, world-class museums, a fantastic food scene, and passionate sports culture.
> Public transport is decent, and many areas are walkable.
> The social scene is lively, and it's easier to build a community than in some other large US cities.
> Safety varies by neighborhood, but generally improving in core areas.
> Key Strengths: Job opportunities, affordability (relative for US), culture, food, public transport (for US).
>
> **7. Montreal, Canada**
> Why it's a great fit: Montreal offers a unique blend of North American dynamism and European charm.
> It's significantly more affordable than Toronto or Vancouver, making it very attractive for recent graduates.
> The city is bilingual (French and English), which adds to its cultural richness.
> It has a thriving arts and music scene, excellent food, and a vibrant nightlife.
> Job opportunities exist, particularly in tech, gaming, aerospace, and creative industries.
> Public transport is efficient, and the city is very walkable.
> It has a strong community feel, especially within its distinct neighborhoods.
> Key Strengths: Affordability (relative for NA), culture, food, social life, unique vibe, public transport.
>
> **8. Singapore**
> Why it's a great fit: If career growth, safety, and efficiency are top priorities, Singapore is hard to beat.
> It's a global hub for finance, tech, and logistics, offering abundant high-paying job opportunities.
> The public transportation system is world-class, and the city is incredibly safe and clean.
> Food is a national obsession, with hawker centers offering delicious and affordable meals.
> There's a large expat community.
> The main downsides are the extremely high cost of living (especially housing and cars) and a potentially more structured social scene compared to some Western cities.
> Key Strengths: Job opportunities, safety, public transport, food, efficiency, cleanliness.
>
> **9. Barcelona, Spain**
> Why it's a great fit: Barcelona offers an incredible lifestyle with beautiful architecture, fantastic food, a vibrant nightlife, and proximity to beaches and mountains.
> It's very walkable, with good public transport.
> The social scene is lively, and there's a strong sense of community.
> While the cost of living is rising, it's still relatively affordable compared to Northern European capitals.
> The main challenge for a recent grad might be job opportunities, especially if you don't speak Spanish or Catalan, and salaries can be lower.
> However, it's a great place for cultural immersion and enjoying life.
> Key Strengths: Lifestyle, culture, food, social life, walkability, climate.
>
> **10. Seoul, South Korea**
> Why it's a great fit: Seoul is a dynamic, high-tech metropolis with an incredibly efficient public transport system, amazing food, and a vibrant youth culture.
> It's very safe, and there's always something happening, from K-pop concerts to traditional markets.
> Job opportunities are strong in tech, media, and design, especially if you have specialized skills or Korean language proficiency.
> It offers a unique cultural experience and a fast-paced environment.
> The main challenges can be the language barrier if you don't speak Korean, and the job market can be competitive for foreigners without specific in-demand skills or a strong network.
> Key Strengths: Culture, food, safety, public transport, technology, dynamism.
>

> ### Which cities might match someone who grew up in Taipei?
> Someone who grew up in Taipei would likely appreciate cities that share some of its characteristics:
> Efficient and extensive public transportation: Taipei's MRT is excellent.
> High safety: Taipei is very safe.
> Vibrant and diverse food culture: Street food and diverse dining options are key.
> High density and walkability: A compact, easy-to-navigate urban environment.
> Strong work ethic and modern infrastructure: A sense of progress and efficiency.
> Cleanliness and order (relative to other cities).
>
> **Cities that might resonate with a Taipei background:**
> 1. Singapore: This is probably the closest match.
> It shares Taipei's hyper-efficiency, incredible safety, world-class public transport, and a fervent food culture (especially hawker centers).
> It's also very clean and modern, with a strong emphasis on career progression.
> 2. Seoul, South Korea: Shares the high-tech integration, incredibly efficient public transport, high safety, and a dynamic, youth-driven culture with fantastic food.
> The urban density and constant activity would feel familiar.
> 3. Amsterdam, Netherlands: While culturally very different, it shares the excellent public transport, high safety, and a very bike-friendly, dense urban environment where things generally work efficiently.
> 4. Melbourne, Australia: While not as dense as Taipei, it has a strong food culture, excellent public transport (trams!), and a high quality of life with a sense of order and cleanliness.
> ---

> ### Which cities feel closer to Philadelphia?
> Someone who studied in Philadelphia might appreciate cities that share some of its characteristics:
> Distinct neighborhoods with unique identities: Philly is known for its diverse neighborhoods.
> Historical significance and a sense of "grit" or authenticity: A less polished, more real urban feel.
> Strong sports culture: Philly fans are passionate.
> Walkable areas but also reliance on cars in some parts: A mix of urban and suburban accessibility.
> A strong sense of local community within a larger city.
> North American urban structure and lifestyle.
>
> **Cities that might feel closer to Philadelphia:**
> 1. Chicago, USA: This is likely the closest match.
> Chicago shares Philly's distinct neighborhoods, strong sports culture, impressive architecture, and a similar "big city, but not overwhelming" feel compared to NYC.
> It also offers a similar blend of historical significance and modern urban life, with a strong sense of local identity.
> 2. Montreal, Canada: While bilingual, Montreal shares a similar historical depth, distinct neighborhood feel, and a strong sense of local culture.
> It has a more European flair but still operates within a North American context, offering a unique blend that might resonate.
> 3. Toronto, Canada: As a major North American city, Toronto shares the diverse population, a strong economy, and a similar urban structure to many US cities.
> While perhaps less "gritty" than Philly, its diverse neighborhoods and strong community feel would be familiar.

> I hope this detailed breakdown helps you narrow down your choices!
> Each city offers a unique experience, and your personal priorities will ultimately guide your decision.
> Good luck with your graduation and your next big adventure!

## **Reflection:**

Reading the AI’s ranking was actually pretty fun. It did a good job breaking down each city in terms of things that matter in your early twenties, like community, food, job opportunities, and what the overall vibe feels like. I also liked how it compared cities that would feel familiar to someone from Taipei versus someone who has lived in Philadelphia. That part felt surprisingly accurate.

What I didn’t love was how casually it treated affordability. Some of the cities on the list are objectively expensive, and the AI sometimes glosses over that reality. I also wish it had offered a few more unexpected choices instead of sticking mostly to well-known global hubs.

Still, the list helped me realize what I actually care about in a city: safety, good public transportation, a strong food culture, and a sense of community. It also made me think about how much a city influences the version of myself I become. Even though AI can’t tell me exactly where to go after graduation, it did help me organize my thoughts and see the decision in a new way.
