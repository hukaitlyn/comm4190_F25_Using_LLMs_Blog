## **Introduction** 

---

Colors have always felt like a quiet way of describing people. Some remind you of bright mornings, others of soft, muted evenings. Each shade carries emotion and energy in its own way. I wanted to see if AI could capture that by looking at me.

I sent ChatGPT a photo of myself and asked it to create a five-color palette based on what it saw. I wanted to know what colors it would choose to represent me and whether any of them would actually feel right.

---
