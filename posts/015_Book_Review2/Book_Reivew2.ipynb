## **Reflection:**

While reading the next section of A Brief History of Intelligence, I started paying more attention to the way Bennett organizes the book. He describes the evolution of intelligence through five big breakthroughs: the first nervous systems, the emergence of learning and memory, the development of flexible behavior through multilayered brains, the rise of language and symbolic reasoning, and finally the appearance of abstract cognition. It is a clean structure that makes the progression feel intentional and easy to follow.

Even so, I found myself grouping the first three breakthroughs together as one larger theme. They all focus on intelligence as something rooted in biology and survival: basic neural circuits, the ability to learn from experience, and behaviors that adapt to changing environments. Thinking about them this way made the early chapters feel connected and helped me understand why Bennett spends so much time on nonhuman species before getting anywhere near humans.

The chapters on animal cognition were especially engaging. Bennett explains how early nervous systems grew from simple loops into networks capable of memory, learning, and flexible behavior. He spends a lot of time on rats, using them as a model for how scientists study learning and decision making. Through maze experiments and reward based tasks, he shows how rats do more than follow rigid habits. They update their behavior when the rules or rewards change, which illustrates a more flexible form of intelligence than simple conditioning. One detail that stuck with me was his description of how rats can form internal “maps” of their environment and use those maps to choose new routes when their usual path is blocked. It is a clear example of how even small animals can build structured representations of the world.

I especially appreciated this perspective as someone studying neuroscience. His explanations felt familiar in some ways, but they also reframed concepts I have learned in class by placing them in a much longer evolutionary timeline. It was interesting to see how he draws connections between biological cognition and the design of AI systems. Even though artificial intelligence does not evolve the way organisms do, Bennett shows how both rely on layered processing, pattern recognition, and forms of feedback that gradually build complexity. The parallel made the material feel relevant to both sides of my academic interests.

On a personal level, these chapters reminded me why cognition feels so interesting to study. It is helpful to see intelligence described as something gradual and adaptive rather than something tied only to language or abstract reasoning. Placing humans within a much longer evolutionary story makes the later discussions about AI feel more grounded and connected to biology instead of separate from it.
