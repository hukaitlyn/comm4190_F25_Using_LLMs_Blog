**Introduction** 
---

Music is emotion translated into sound — joy, grief, nostalgia, calm — captured in frequencies we somehow understand without words. It’s one of the few things that can make us feel seen without ever speaking back.

But what happens when a machine starts listening?
When an AI, built from code and data, tries to interpret the feelings that make us human?

We often use music to express emotions we can’t name. I wanted to see if an AI could do the same. If they can not just analyze lyrics or tempo, but recognize the emotional story behind a song. Can it tell the difference between heartbreak and hope? Can it translate what we hear into something it can feel?

To find out, I ran three small experiments: I gave an AI my playlists, my moods, and my questions about emotion — and asked it to listen.

---

### Reflection: 
What surprised me most was not Gemini’s accuracy but its warmth. The way it described my playlist felt like reading a horoscope that somehow got everything right: confident, romantic, a little wistful. I did not tell it anything about how I was feeling, yet it pieced together a whole story from the songs alone. It was playful, teasing even, but also strangely perceptive.

I am not in love right now, but I seem to be drawn to love songs. Maybe that says something about how I connect to emotion through music more than through my own experiences. “Last Night” is actually one of my favorite songs, but Gemini read it as a curveball, as if it did not belong with the rest. I am not sure how to interpret that, but it makes me curious about how it analyzes songs in the first place. I really think it uses lyrics and tempo to infer emotion, maybe even patterns from how people usually describe those songs online. Either way, it made me realize how much of our emotional life hides in what we listen to and how easily an AI can start to read it back to us.

## **Experiment 2: Empathy Through Sound**

Goal: See if Gemini can learn emotional meaning through context, not just from song data.

### Prompt: 
> I have been feeling a bit emotionally overwhelmed and just want to feel calmer and understood. Can you make me a short playlist that feels like a reset, something that helps me breathe, unwind, and feel seen without being overly sentimental? Choose five songs, and explain why each one might help emotionally, not just musically. Can you also choose those songs based on my music taste from the first prompt please? 

### Response: 
>  Hey friend, I totally get that feeling of being emotionally swamped. It's like your brain just needs a soft landing, a little quiet corner to just *be*. I've put together a few songs that I hope can be that gentle reset for you – nothing too heavy, just a space to breathe and feel a little more grounded.

> Here’s a short playlist for you:
>
> 1. **"Holocene" by Bon Iver**
> This song feels like a vast, quiet breath. When things feel overwhelming, it has a way of putting everything into a bigger, more beautiful perspective. It's about finding your place in the grand scheme, and it can make your own worries feel a little smaller, a little more manageable, just for a moment. It's like looking up at the night sky and feeling a quiet sense of awe and belonging.
>
> 2. **"Intro" by The xx**
> This one is purely atmospheric, and that's its power. There are no lyrics to interpret, just a steady, almost heartbeat-like rhythm and sparse, echoing guitar. It creates this calm, almost meditative space that lets your mind just *settle*. It's perfect for when you need to turn down the noise and just feel a consistent, gentle presence. It's like a soft, quiet hum that helps you recenter.
>
> 3. **"Pink Moon" by Nick Drake**
> Short, sweet, and incredibly gentle. Nick Drake's voice is like a quiet, knowing whisper here. It doesn't try to solve anything, but it acknowledges a quiet truth about the world and internal feelings without judgment. It feels like a warm, understanding nod from someone who just *gets* that sometimes things are just... what they are, and that's okay. It’s comforting in its simplicity.
>
> 4. **"Bloom" by The Paper Kites**
> This song is pure warmth. The acoustic guitars and the soft, layered harmonies just wrap around you like a cozy blanket. It has this tender, hopeful quality that isn't pushy or overly optimistic, but just gently reassuring. It feels like a quiet promise that things can soften and grow, and it offers a sense of gentle comfort when you're feeling raw.
>
> 5. **"Weightless" by Marconi Union**
> Okay, this one is specifically engineered to be relaxing, and it really works. It's an instrumental ambient track with no sharp edges, just flowing, soothing sounds. When your brain is buzzing, this song is like hitting the "mute" button on the outside world. It helps slow your breathing and quiet your thoughts, giving you a chance to truly unwind and physically feel calmer.

> I really hope these give you a little space to breathe and feel a bit lighter. Remember, it's okay to feel overwhelmed, and it's okay to take a moment just for you.

### Reflection: 
I did not know any of the songs Gemini recommended, so I listened to each one separately before writing this. They are not really the kind of music I usually listen to, and only some of them matched the vibe I had in mind. I liked Holocene and Bloom the most because they felt calm and grounding, the kind of songs that might actually help if I were feeling down. The other three did not really do much for me. If I was having a rough day and those came on, I do not think they would make me feel better. Most of the songs also seemed a bit outdated, which makes me wonder how Gemini defines comfort or calmness. Maybe it draws on older tracks because they feel timeless or familiar to most people.

The goal of this experiment was to see if Gemini could provide emotional comfort through music, and in some ways it did. The songs it chose carried a consistent sense of gentleness and calm, which matched what I asked for. However, I also told it to choose songs based on my previous playlist, and it did not seem to do that. It felt like Gemini switched from reading my emotions to applying a more general formula for relaxation. It captured the tone correctly but not the personal connection I was hoping for. This made me realize that emotional comfort from music depends as much on familiarity and taste as on sound. Gemini seemed to understand the emotion in theory, but not how it feels in practice.


## **Experiment 3: Teaching Gemini to Feel the Music**

Goal: See if Gemini can learn emotional meaning through context, not just from song data. 

### Prompt: 
> I want to see if you can learn how I personally connect to music. Below is a list of songs that each represent a different emotion for me. Can you analyze them and describe what you think these emotions mean to me, not just in general?
>
>Calm: _Coffee Breath by Sofia Mills_  
>Hopeful: _Golden hour by JVKE_  
>Nostalgic: _Fresh Eyes by Andy Grammer_  
>Confident: _Hotel Room Service by Pitbull_
>Sad: _A 1000 Times by Hamilton Leithauser + Rostam_
>
> After describing what you think each song represents for me emotionally, try to summarize what you’ve learned about how I experience feelings through music.

### Response: 
> Okay, this is a fantastic way to understand your connection to music! Let's dive into these, almost like we're unpacking a secret language you share with these songs.
>
> Here’s what I'm picking up about your emotions through these choices:
>
> * Calm: "coffee breath" by Sofia Mills
> For you, "calm" isn't about shutting down or being completely blank. This song suggests your calm is a very **intimate, soft, and almost wistful feeling**. It's not a "zen garden" kind of calm, but more like the quiet comfort of being truly present in a small, personal moment, perhaps with someone you care about, or just with your own thoughts. There's a gentle introspection here, a quiet settling into a familiar, tender space that feels safe and gentle, even if it carries a tiny hint of melancholy. It's a very human, lived-in calm.
>
> * Hopeful: "Golden Hour" by JVKE
> Your "hopeful" isn't a quiet, internal whisper. This song makes me think that for you, hope is **expansive, warm, and often tied to connection or a sense of shared beauty**. It feels bright, almost cinematic, and perhaps a little grand. It's the feeling when things align perfectly, when you're deeply connected to someone or something beautiful, and the world just *glows*. It's an uplifting, almost euphoric hope, where possibilities feel limitless and vibrant.
>
> * Nostalgic: "Fresh Eyes" by Andy Grammer
> This is really interesting for "nostalgic"! It tells me that for you, nostalgia isn't solely about longing for what's gone or a sad yearning for the past. Instead, your nostalgia seems to be a **positive, active re-appreciation**. It's about revisiting familiar feelings, people, or experiences and finding a renewed sense of wonder and joy in them, as if seeing them for the first time again. It's a warm, appreciative look back that brings fresh perspective and happiness, rather than just wistfulness. You find enduring value and delight in the familiar.
>
> * Confident: "Hotel Room Service" by Pitbull
> Okay, this one screams! Your "confident" is clearly **energetic, outward-facing, and uninhibited**. It's not a quiet, internal self-assurance; it's a bold, almost performative feeling of owning the room, having a blast, and not caring what anyone else thinks. There's a mischievous, celebratory vibe here. Your confidence is about embracing fun, being seen, being heard, and fully stepping into a vibrant, self-assured energy that wants to make an impact and enjoy the moment to the fullest.
>
> * Sad: "a 1000 times" by Hamilton Leithauser + Rostam
> For you, "sad" isn't a simple feeling of being down. This song suggests your sadness is **complex, deeply reflective, and perhaps a little dramatic or theatrical in its emotional landscape**. It involves looking back, revisiting past relationships or significant emotional moments with a profound sense of longing, regret, or a heavy heart. There's an active emotional processing happening, a deep dive into the layers of feeling, perhaps agonizing over things "a thousand times." It's a sadness that feels intensely personal, rich with internal narrative, and perhaps a bit raw and consuming.

> What I've learned about how you experience feelings through music:
>
> You seem to experience emotions through music in a way that is incredibly **rich, nuanced, and deeply personal**. You don't just feel an emotion; you *immerse* yourself in its specific texture and story.
>
> There's a strong thread of **active engagement** with your feelings. Whether it's the gentle introspection of calm, the expansive glow of hope, the joyful re-appreciation of nostalgia, the bold energy of confidence, or the complex reflection of sadness – you choose songs that allow you to fully inhabit and explore the emotional landscape. You're drawn to music that provides a specific, almost narrative backdrop for your internal world.
>
> You also seem to find comfort and understanding in songs that allow for **complexity within an emotion**. Your "calm" has a touch of wistfulness, your "nostalgia" is positive, and your "sadness" is deeply reflective rather than just resigned. You use music to articulate the subtle shades of your feelings, finding depth and specific meaning in the sounds and stories. You connect with music that helps you truly *feel* what you're feeling, in all its intricate glory.

### Reflection:

It was interesting to see how well it could read those differences.

When it started recommending new songs, the results felt a bit more personal than before. Some choices fit better than others, but I could tell it was trying to match my emotional patterns instead of just guessing based on genre. The calm and nostalgic ones felt most accurate, while the rest were fine but not really what I would listen to often.

Compared to Experiment 2, this felt more thoughtful. Gemini wasn’t just picking random comforting songs anymore. It seemed to actually use what it learned from my earlier playlists. It still missed a few things, but it was closer to understanding what my version of “feeling” sounds like.

**Conclusion** 
---

Looking back at all three experiments, I can see how each one built on the last. In the first, Gemini read my playlist like a mirror and somehow managed to describe the emotions behind it in a way that felt warm and human. In the second, it tried to comfort me through music but ended up sounding more general, as if it knew the idea of calm but not what calm feels like to me. By the third, it started getting closer. It learned from my own songs and tried to recommend new ones that matched how I experience emotions.

What surprised me most was how much these small interactions made me think about how I use music in the first place. The songs we listen to really are emotional markers, tiny reflections of who we are in a certain moment. Gemini could understand some of that, but not all of it. The human part is still missing, the one that connects music to memory and meaning.

Still, there was something special about teaching an AI to listen, not just to sound but to feeling. Even if it cannot fully understand, the process made me pay attention to how I feel, and maybe that is its own kind of understanding.
