Touch rugby is one of those sports most people in the United States have never heard of, but the rest of the world absolutely loves. It is especially big in Australia and New Zealand, where people grow up playing it the way kids here grow up playing basketball or soccer. The easiest way to describe it is that it is a version of rugby without tackling. Instead of getting taken to the ground, play stops with a simple touch. Think of it as tag with a ball, except everyone is sprinting, passing, stepping, and trying to outsmart each other at full speed. It looks simple until you try it, and then you realize how technical and exhausting it actually is.

In two weeks, I will be attending a three-day touch rugby training camp as part of my preparation for USA Nationals with the NYC team. The camp runs from eight in the morning to four in the afternoon and each day is packed with drills, attacking lines, defensive patterns, footwork, conditioning, sprint work, and constant ball handling. It will be the most intense stretch of touch rugby training I have done so far, and I want to make sure I am actually ready for it.

Two weeks is not a lot of time to prepare, and I did not want to show up to camp feeling unprepared. So instead of randomly picking workouts or hoping practices would be enough, I asked an LLM to help me create an actual training plan. I wanted something realistic that matched the speed, movement, and ball-handling demands of touch rugby. This post includes the prompt I used, the plan the AI created, and my first impressions before I try it out.

## Reflection 

Even though I already know how to train for touch rugby, it was still interesting to see how the AI organized everything. The plan felt structured and sport specific, and it brought together elements I already focus on, like sprint work, agility, ball handling, endurance, and recovery, but in a different format than I usually create for myself. It did not teach me anything completely new, but it did lay everything out in a way that made the next two weeks feel more intentional and easy to follow.

I was surprised that the AI did not ask me any follow-up questions about my schedule, my current level of athleticism, or whether I had any injuries. A real coach would have asked those things immediately, so the lack of personalization made the response feel less human and more template-based. Even so, some of the workouts were genuinely interesting and more creative than what I would have come up with on my own. It reminded me that AI can give strong starting points, even if it cannot fully tailor a program the way a live coach could.

Overall, I can see myself using this kind of plan again for inspiration or structure, especially when I am short on time or want ideas for specific types of drills. It is not a replacement for a touch coach, but it is definitely a useful tool for staying organized, motivated, and a little more creative with training.
