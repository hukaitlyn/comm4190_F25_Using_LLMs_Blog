## **Introduction:**

My roommate’s birthday is coming up, and I wanted to plan something small and thoughtful without overthinking it. Instead of scrolling through ideas or asking everyone I know, I decided to try using an LLM to help me come up with a simple plan. I wanted to see whether AI could suggest something that actually fits our personalities, our budget, and the way we usually spend time together.

I gave the model a short description of my roommate and asked it to design a birthday surprise that feels personal but easy to put together in our apartment. The next sections show the prompt I used, the information I provided, and the plan the AI created.

## **Prompt:** 

> You are an event planner helping me create a thoughtful birthday surprise for my roommate. Use only the information I give you. Create a complete plan that fits a college budget and feels personal and realistic. Include activity ideas, food suggestions, small gift options, and simple preparation steps. Include one main idea and one or two backup options. At the end, ask if I want anything changed.

### What I Told AI About My Roommate: 

> Roommate’s vibe: Calm, friendly, and easygoing.
>
> Interests: Sweet snacks, cozy nights in, skincare, and simple gifts.
>
> Dislikes: Loud events, big crowds, and anything that feels overly planned.
>
> Budget: A moderate college budget.
> 
> Type of surprise I want: A small, meaningful celebration we can do at home.
>
> Birthday month: October.

## **Response:**

> Here's a thoughtful birthday surprise plan for your roommate, designed to be personal, realistic, and fit a college budget, using only the information you provided:
>
> Your roommate's calm, friendly, and easygoing vibe, coupled with their love for cozy nights in, sweet snacks, skincare, and simple gifts (and dislike for anything loud, crowded, or overly planned), perfectly sets the stage for a relaxed at-home celebration. Since their birthday is in October, we'll lean into a cozy, autumn-inspired theme.
>
> ---
>
> ### **Main Idea: Cozy Autumn Glow-Up & Movie Night**
>
> This plan combines their love for skincare and cozy nights into one relaxing, personal experience.
>
> **Activity Ideas:**
> 1.  **DIY Spa & Movie Marathon:** Casually suggest a "let's just relax tonight" vibe. Start with a mini at-home spa session. You can do simple, budget-friendly **sheet masks** together (look for multi-packs at drugstores or beauty supply stores, often $2-5 each) or even make a quick **DIY face mask** using common kitchen ingredients like honey and oatmeal. Put on some chill, calm background music while you relax with your masks.
> 2.  **Cozy Film Choice:** Transition into a movie night. Let your roommate pick the film – something comforting, not too intense, and perfect for a relaxed evening in. Cuddle up with blankets and pillows.
>
> **Food Suggestions (Sweet Snacks Focus):**
> *   **Homemade Cookies:** Nothing says "cozy" like fresh-baked goods. Whip up a batch of their favorite cookies (chocolate chip, oatmeal, or even a simple sugar cookie). This is very budget-friendly and smells amazing!
> *   **Gourmet Chocolate Bar:** Pick up one or two higher-quality chocolate bars (their favorite flavor, perhaps with an autumnal twist like sea salt caramel or dark chocolate orange).
> *   **Warm Apple Cider/Hot Cocoa Bar:** Since it's October, warm apple cider is perfect. You can buy a jug and warm it on the stove or microwave. Alternatively, set up a simple hot cocoa bar with cocoa powder, milk, marshmallows, and maybe a sprinkle of cinnamon.
> *   **Apple Slices with Caramel Dip:** A simple, sweet, and seasonal treat.
>
> **Small Gift Options:**
> *   **Skincare Essentials:** A couple of nice, budget-friendly **sheet masks** (if you didn't use them for the activity), a good quality **lip balm**, or a small, hydrating **hand cream**.
> *   **Cozy Comforts:** A pair of soft, **fuzzy socks** (especially good for autumn!), a small **fall-scented candle** (pumpkin spice, apple cinnamon, or vanilla bean), or a cute, simple **mug** for their warm drinks.
> *   **Thoughtful & Simple:** A small, aesthetically pleasing **notebook** and a nice pen, or a comforting **book** from a thrift store that aligns with their interests.
>
> **Simple Preparation Steps:**
> 1.  **The Day Before/Morning Of:** Pick up any ingredients for baking, drinks, and gifts.
> 2.  **Before They Get Home:** Tidy the common area (living room or your shared dorm space). Arrange blankets and pillows on the couch or bed to create a super cozy nook. Dim the lights, maybe use fairy lights if you have them, or just rely on lamps for a soft glow.
> 3.  **Food Prep:** Bake the cookies or arrange store-bought snacks on a nice plate or board. Prepare the hot cocoa ingredients or warm the apple cider.
> 4.  **Gift Placement:** Place the chosen small gift(s) neatly on a side table near the cozy setup, perhaps with a simple, heartfelt handwritten card.
> 5.  **The Reveal:** When they arrive, casually say something like, "Hey, I thought we could have a really chill night in tonight to celebrate your birthday!" and reveal the cozy setup, snacks, and gifts. Make it feel relaxed and spontaneous rather than a rigid schedule.
>
> ---
>
> ### **Backup Option 1: Sweet Treat Board & Chill Music**
>
> If the skincare element feels like too much, this option focuses purely on their love for sweet snacks and a relaxed atmosphere.
>
> **Activity Ideas:**
> 1.  **Curated Playlist & Hangout:** Create a personalized playlist of calm, friendly music that you know your roommate enjoys. Let it play softly in the background.
> 2.  **Casual Chat or Companionable Silence:** Just hang out, chat, or enjoy each other's company in comfortable silence while munching on treats. This is perfect for someone who dislikes "overly planned" events.
>
> **Food Suggestions (More Sweet Snacks!):**
> *   **Dessert Platter:** Focus on a beautiful spread of their favorite sweet snacks. This could include homemade cookies (as above), a small box of gourmet chocolates, a couple of their favorite candy bars, and maybe a small personal-sized cake or a few cupcakes from a local bakery.
> *   **Warm Drinks:** Hot cocoa, herbal teas, or warm apple cider.
>
> **Small Gift Options:**
> *   **Curated Candy Box:** Assemble a small box or bag of their absolute favorite candies and chocolates.
> *   **Cozy Mug:** A unique or aesthetically pleasing mug for their warm drinks.
> *   **Comfort Item:** A nice, soft throw blanket (if budget allows for a smaller one) or a really good quality, comforting herbal tea blend.
>
> **Simple Preparation Steps:**
> 1.  **Before They Get Home:** Tidy the space. Set up a dedicated "snack zone" on a coffee table or desk, making it look inviting.
> 2.  **Food Prep:** Arrange all the sweet snacks attractively on a board, platter, or in small bowls. Prepare the drinks.
> 3.  **Gift Placement:** Place the small gift next to the snack board with a heartfelt birthday card.
> 4.  **The Reveal:** Have the music playing softly. When they arrive, simply say, "Surprise! I thought we could just chill out with some treats and good music for your birthday."
>
> ---
>
> I hope these ideas help you create a wonderfully thoughtful and personal surprise for your roommate!

There were a few parts of the plan that I genuinely liked. The cozy at home theme matched my roommate’s personality well, and the suggestions for snacks, warm drinks, and a simple movie night felt realistic for how we already spend time together. I also liked the small gift ideas since they stayed within a college budget and were things my roommate would actually use. The AI did a good job keeping everything easy to prepare and not overwhelming.

There were also parts that I did not love. The tone was a little more dramatic and enthusiastic than how I would normally talk about a birthday surprise. Some of the long explanations felt unnecessary, and certain suggestions were a bit too detailed for what I was looking for. I also would have preferred fewer steps and a shorter list overall.

Even so, the plan was helpful because it gave me a clear starting point. It made the surprise feel manageable, and it highlighted a few ideas I would not have thought of on my own.
