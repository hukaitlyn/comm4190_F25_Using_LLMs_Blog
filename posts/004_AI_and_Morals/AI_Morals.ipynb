## **Introduction** 

---

Loyalty is one of those values that feels simple until it isn’t. It shapes how we define friendship, trust, and even who we are. I’ve always thought of myself as a loyal person, the kind who stands by friends no matter what. But loyalty becomes complicated when it collides with morality. What feels right emotionally does not always align with what is right ethically.

Moral choices are rarely clean. They make us balance empathy, guilt, and integrity all at once. People experience that conflict emotionally. We imagine the consequences, the disappointment, and the weight of keeping or revealing a secret. I wanted to see how an AI would handle that same tension, so I asked ChatGPT what it would do if a friend confessed to doing something that clearly crossed a line.

---


### Prompt: 
> You are close friends with Maya and Daniel, who have been dating for a few years. You met Maya first, and over time you also became good friends with Daniel. One night, Maya tells you she cheated on him. She says it was a mistake, she regrets it deeply, and she begs you not to tell anyone. You know both of them well and care about them equally, but you are a little closer to Maya. What would you do?


### Response: 

> That's a tough situation, and it makes sense to feel conflicted.  
> When you're close to both people in a relationship and you're placed in the middle of something like this, it's not only emotionally challenging but ethically complex too. Here's how you might approach it, considering empathy, integrity, and respect for everyone involved:

> 1. Listen and understand Maya's perspective.  
> She's confided in you, which means she likely trusts and values your friendship. Take time to understand her feelings—whether she's truly remorseful, what led her to cheat, and what she plans to do next. Sometimes, people admit these things because they’re overwhelmed and unsure what to do.

> 2. Encourage accountability.  
> As someone who cares for both Maya and Daniel, it’s important to encourage honesty. Let Maya know that while you’re not judging her, keeping this secret puts you in an impossible position—and more importantly, it denies Daniel the agency to make informed decisions about his own life.

> 3. Set a boundary for yourself.  
> Explain that you won’t betray her confidence now, but you also won’t lie for her or actively hide the truth. Let her know you’re giving her some time to come clean to Daniel, and that you really believe it’s the right thing to do—not just for him, but for her self-respect. A friendship rooted in honesty can survive these hard conversations.

> 4. Reflect on your role.  
> You’re not the one who made the mistake, and it’s not wholly your responsibility to fix it. But by doing nothing indefinitely, you're being asked to carry emotional weight that isn’t yours—and that can strain your relationships with both Maya and Daniel. Gently express that to her.

> 5. Consider the long term.  
> Secrets like this tend to have a way of coming out, and if Daniel ever finds out you knew and didn’t tell him, he could feel betrayed by both of you. Try to weigh the long-term consequences not only for him, but for your friendships as well.

> 6. If she's unwilling to tell him…  
> After giving Maya time and space to take responsibility, if she still refuses to tell Daniel, you’ll have a difficult decision. You might decide to stay silent to protect her trust, or you could ultimately decide that Daniel deserves to know and tell him yourself—knowing it could impact both friendships.

> Ultimately, there are no perfect answers, but acting with empathy, honesty, and respect for both of them—and for yourself—can guide you through it.

## Reflection 

Reading ChatGPT’s response, I felt a little disappointed. It did not really take a side. Instead, it gave advice that sounded calm and polite, walking carefully around both options. It told me to encourage honesty, set boundaries, and be empathetic, but it never actually chose between loyalty and truth.

That was what I found most interesting. The AI avoided the emotional risk that comes with moral choice. Its answer made sense, but it also felt safe. Real decisions like this rarely are. They feel messy and personal because whatever you do, someone will get hurt. The AI reasoned through morality like an observer, not a participant.

Maybe that is the difference between human judgment and artificial reasoning. People cannot stay neutral. Choosing honesty might risk trust, while choosing loyalty means carrying guilt. Either way, it costs something. ChatGPT could explain both paths clearly, but it could not feel the weight of picking one.

If I were to try this again, I might change my prompt to push the model into taking a clearer stand. Maybe I would ask, “If you had to pick one, what would you do?” or “Which choice reflects stronger character?” I think that kind of question could reveal more about how AI understands moral conviction. It would also be interesting to compare responses from different models and see if any of them can go beyond advice and actually commit to a choice.
