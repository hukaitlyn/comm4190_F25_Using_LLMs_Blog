## **Introduction:**

Halloween and I have a long history of doing the absolute minimum. Most years I panic grab something from my closet and hope no one notices I put in zero effort. But since it is my senior year, I wanted to at least try a little harder. Not full Pinterest dedication, just better than usual. My goal was to reuse as much as possible from my closet while still looking cute, fun, and a little creative.

Instead of spending hours scrolling or overthinking it, I decided to let an LLM choose my Halloween costume for me. I wanted to see if AI could actually understand my vibe and give me ideas I would genuinely wear. Nothing expensive, nothing uncomfortable, just realistic options that match my style. I was not sure if it would be helpful or chaotic, but it made the whole decision feel more fun and a lot less stressful.

## **Prompt:** 

> I need help choosing my Halloween costume this year. I am a college senior and I want to try a little harder than usual while still using as much as I can from my own closet. I want something cute, fun, and creative but still realistic and comfortable to wear. Ask me three short questions first to understand my style, personality, comfort level, and what I might already own. After I answer, give me at least five costume ideas that genuinely fit my vibe. Explain why each idea matches me and how easy it would be to put together with common clothes and accessories.

### Getting To Know My Style: 

> **ASSISTANT:**
> Okay, I can help you brainstorm some costume ideas that fit your style. Answer these three quick questions.
>
> 1. Describe your everyday style in three words. What colors do you usually wear
> 2. On a scale from 1 to 5, how much effort are you willing to put into a costume, and are you okay with simple makeup or props
> 3. Is there a theme, aesthetic, or show you have been drawn to lately You can keep it general
>
> I will give you personalized ideas after your answers.

> **USER:**
>
> 1. comfy, trying to be trendy, mostly neutrals like black, white, grey, navy
> 2. normally a three, but for Halloween a four
> 3. anything cute and recognizable, nothing anime

## **Reflection:**

Trying this experiment actually ended up being more helpful than I expected. I thought the AI would give me random or over-the-top ideas, but the suggestions were surprisingly on point. What made me laugh a little was how many of the characters it mentioned were ones I had already brainstormed myself at some point. Wednesday Addams, Audrey Hepburn, a cute vampire moment, even Reputation era Taylor Swift all fit the exact color palette I wear every day. It made me feel like the AI actually understood my style better than I expected.

I also appreciated that all the ideas were realistic and comfortable, which is exactly what I wanted. Nothing felt dramatic or out of character for me. Everything was cute, recognizable, and easy to build from pieces I already own. It felt like the AI filtered my entire closet and personality into a small list of options that made total sense.

Overall, I was pleasantly surprised by how well this worked. It made the whole process feel more fun and way less stressful, and it reminded me that my everyday style already leans toward a lot of Halloween-adjacent looks. I would definitely use AI again for costume brainstorming, especially when I want something low effort but still cute.
