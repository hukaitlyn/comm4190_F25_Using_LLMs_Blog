## **Introduction** 

---

Every friend group has unspoken roles—the leader, the calm one, the funny one, the planner. I wanted to see how an AI would interpret those same dynamics from the outside. I uploaded a photo of my friends and me to ChatGPT-5 and asked it to describe what it saw. I was curious whether it could recognize any kind of relationship pattern or if it would simply make confident guesses based on our poses and expressions.

When people look at a picture, we instantly read mood and connection: who feels close, who looks distracted, who’s having fun. But when a machine does the same, is it seeing friendship or just the visual data that resembles it?

---

## Reflection 
When I read the AI’s breakdown, I was surprised by how confident it sounded, like it really knew us. At first, I thought its guesses were random, but some of them actually made sense. The person it called “the leader” was standing in the middle, slightly forward, and wearing a brighter outfit. That probably did make her stand out more, and in psychology, people who take up space or hold the center position are often seen as more dominant or in charge. The friend it labeled “the funny one” had her arms around everyone and was smiling the widest. That kind of body language is usually linked to being outgoing and sociable.

Even though the AI doesn’t know us, it was picking up on real nonverbal signals, things humans also read without thinking. Psychologists call this “thin-slice perception,” and it explains how we form quick impressions from small visual cues. The difference is that people know their impressions might be wrong, while the AI treats them as facts.

Looking at its results made me realize how much body language shapes the stories we tell about people. In some ways, the AI wasn’t completely off. The calm posture, the open smile, the center of attention—it all reflects something real, but not the whole truth. It could see the patterns, but it couldn’t see the friendships behind them.
